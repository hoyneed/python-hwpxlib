```python
import subprocess
from pathlib import Path

hwpx_jar_path = Path('./hwpxlib-1.0.5.jar')
hwpx_file = '2023년 디지털정부 발전유공 포상 추진계획.hwpx'

text = subprocess.run(["python", "hwpx_loader.py", "--hwpx_jar_path", hwpx_jar_path, "--file_path", hwpx_file],
                      capture_output=True, text=True)
```

```python
print(text)
```

```python
import requests

url = "http://localhost:7860/extract-text"
file_path = "2023년 디지털정부 발전유공 포상 추진계획.hwpx"

with open(file_path, 'rb') as f:
    files = {'file': (file_path, f)}
    response = requests.post(url, files=files)

print(response.json())
```
